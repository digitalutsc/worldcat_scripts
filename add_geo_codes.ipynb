```python
from urllib.request import urlopen
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
from geopy.geocoders import Nominatim 
from geopy.exc import GeocoderTimedOut
import glob
from geopy import GoogleV3
import time
```

```python
# Cache info into a list so we can avoid duplicate calls
institutions_info = pd.read_csv("data/institutions_info.csv")
institutions_info.set_index('institutionIdentifier', inplace=True)
    
def check_if_in_org_list(institutions_info, institutionIdentifier):
    if institutionIdentifier in institutions_info.index:
        row = institutions_info.loc[institutionIdentifier]
        return row
    else:
        return None
    
def add_to_org_list(institutions_info, institutionIdentifier, physicalAddress, latitude, longitude):
    row = pd.Series({'physicalAddress':physicalAddress,'latitude':latitude,'longitude':longitude},name=institutionIdentifier)
    institutions_info = institutions_info.append(row)
    return institutions_info
```

```python
# Get geocode info from Google if not on list
geolocator = Nominatim(user_agent="your_user_agent")

def do_geocode(physicalAddress):
    try:
        location = GoogleV3(api_key="yourgeocodeapi").geocode(physicalAddress)
        return location
    except GeocoderTimedOut:
        print("re trying")
        time.sleep(10)
        return do_geocode(physicalAddress)
    

def get_geo_code(institutions_info, institutionIdentifier, physicalAddress):
    
    location = check_if_in_org_list(institutions_info, institutionIdentifier)
    if location is not None:
        print("in list")
        return location, institutions_info
    
#    location = geolocator.geocode(physicalAddress)
#    if location is not None:
#        institutions_info = add_to_org_list(institutions_info, institutionIdentifier, physicalAddress, location.latitude, location.longitude)
#        return location, institutions_info
    location = do_geocode(physicalAddress)
    if location is not None:
        print("in google")
        institutions_info = add_to_org_list(institutions_info, institutionIdentifier, physicalAddress, location.latitude, location.longitude)
        return location, institutions_info

    return location, institutions_info
```

```python
# Loop through each file to get geocode info
def get_locations_by_file(file_path, institutions_info):
    holding_info = pd.read_csv(file_path)

    for index, row in holding_info.iterrows():
        institutionIdentifier = str(row["institutionIdentifier"])
        physicalAddress = str(row["physicalAddress"])
        geocodes, institutions_info = get_geo_code(institutions_info, institutionIdentifier, physicalAddress)
        if geocodes is not None:
            holding_info.loc[index, 'latitude'] = geocodes.latitude
            holding_info.loc[index, 'longitude'] = geocodes.longitude

    file_path2 = file_path.replace("holdings", "holdings2")            
    holding_info.to_csv(file_path2)
    return institutions_info
```

```python
csv_files = glob.glob("data/extracted/holdings/*.csv")

for file_path in csv_files:
    print(file_path)
    institutions_info = get_locations_by_file(file_path, institutions_info)
```
